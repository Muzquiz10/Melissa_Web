```python
import qrcode
from qrcode.constants import ERROR_CORRECT_H
from PIL import Image, ImageDraw
import pandas as pd
import os
```

### Creador QR

```python
# -- Variables para el proceso -- #
enlace = "https://muzquiz10.github.io/Melissa_Web/index.html"
logo = Image.open("../assets/Icons/Logo_Melissa.png").convert("RGBA")
output_dir = "qrs"
nombre_archivo = "qr_melissa"
#----------------------------------#
# --- Crear QR --- #
qr = qrcode.QRCode(
    version=None,
    error_correction=ERROR_CORRECT_H,
    box_size=10,
    border=4,
)

qr.add_data(enlace)
qr.make(fit=True)

qr_img = qr.make_image(fill_color="#ae8d76", back_color="white").convert("RGBA")

qr_width, qr_height = qr_img.size

# --- Tamaño del logo (20–22% recomendado) ---
logo_size = int(qr_width * 0.22)
logo_resized = logo.resize((logo_size, logo_size), Image.LANCZOS)

# --- Coordenadas centradas ---
x = (qr_width - logo_size) // 2
y = (qr_height - logo_size) // 2

# --- 1️⃣ Crear hueco blanco en el centro ---
draw = ImageDraw.Draw(qr_img)
padding = int(logo_size * 0.15)  # margen alrededor del logo

draw.ellipse(
    [
        x - padding,
        y - padding,
        x + logo_size + padding,
        y + logo_size + padding
    ],
    fill="white"
)

# --- 2️⃣ Pegar logo ---
qr_img.paste(logo_resized, (x, y), logo_resized)

# --- Guardar ---
qr_img.save(f"{output_dir}/{nombre_archivo}.png")
```
